##What is PCA?
**Principal Component Analysis (PCA)** is a dimensionality reduction technique that is commonly used to reduce the number of dimensions in a dataset while preserving as much variance as possible. It works by transforming the original features into a new set of orthogonal components called principal components. These components are ordered by the amount of variance they explain in the data.

##Why Use PCA?
- Dimensionality Reduction: PCA helps in reducing the number of features (dimensions) in a dataset, which can simplify subsequent analysis and modeling.
- Visualization: PCA can aid in visualizing high-dimensional data in a lower-dimensional space, making it easier to understand and interpret patterns.
- Noise Reduction: PCA can sometimes help in filtering out noise and focusing on the most significant patterns in the data.
- Feature Engineering: PCA can be used for feature engineering by creating new features that capture the most variance in the data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
```

```python
feature_names
```

```python
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled [0:5]
```

```python
# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca[0:5]
```

```python
# Create a DataFrame for visualization
df_pca = pd.DataFrame(data=X_pca, columns=['Principal Component 1', 'Principal Component 2'])
df_pca['Target'] = y
```

```python
# Visualize PCA components
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Principal Component 1', y='Principal Component 2', hue='Target', data=df_pca, palette='Set1', s=100, alpha=0.8)
plt.title('PCA: Iris Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Target', loc='upper right')
plt.grid(True)
plt.show()
```
